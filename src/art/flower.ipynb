```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# --- Canvas Setup ---
size = 600
canvas = np.zeros((size, size))
cx, cy = size // 2, size // 2

# --- 1. Draw the Petals (Sunflower Style) ---
# More petals to look like a sunflower
num_petals = 18 
petal_length = 250

for theta in np.linspace(0, 2 * np.pi, 4000): # Increased points for detail
    # The core formula for the rose shape
    r = petal_length * np.sin(num_petals * theta)
    
    # Convert from polar (r, theta) to cartesian (x, y)
    x = int(cx + r * np.cos(theta))
    y = int(cy + r * np.sin(theta))
    
    # Draw the point on the canvas if it's within bounds
    if 0 <= x < size and 0 <= y < size:
        # Add a "glow" to make the petals look softer
        for dx in range(-3, 4): # Smaller brush for more defined petals
            for dy in range(-3, 4):
                if 0 <= x + dx < size and 0 <= y + dy < size:
                    if dx**2 + dy**2 < 9: # Circular brush
                        # Assign a higher value for yellow
                        canvas[x + dx, y + dy] = 2.0

# --- 2. Draw the Textured Center ---
center_radius = 120 # Larger center for a sunflower
for i in range(size):
    for j in range(size):
        if (i - cx)**2 + (j - cy)**2 < center_radius**2:
            # Add random noise for a seed-like texture
            canvas[i, j] = 0.5 + np.random.rand() * 0.5

# --- 3. Visualization ---
plt.figure(figsize=(10, 10))
sns.heatmap(canvas, cmap='YlOrBr_r', cbar=False)
plt.show()
```
