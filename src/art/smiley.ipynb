# Heatmap art 
## Smiley

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = 200
Z = np.zeros((size, size)) # Initialize a 200x200 grid

# Circle mask
cx, cy = size//2, size//2
for i in range(size):
    for j in range(size):
        if (i-cx)**2 + (j-cy)**2 < (size//2.4)**2: # inside circle
            Z[i, j] = 1   # face (Yellow)

# Eyes
for i in range(size):
    for j in range(size):
        # Left Eye (above center, left)
        if (i-(cx-30))**2 + (j-(cy-35))**2 < 150:
            Z[i, j] = 0   # Eyes (Black)
        # Right Eye (above center, right)
        if (i-(cx-30))**2 + (j-(cy+35))**2 < 150:
            Z[i, j] = 0   # Eyes (Black)

# Smile
# Arc for the smile (below center)
for theta in np.linspace(0.2, np.pi-0.2, 500):
    for r_offset in range(-3, 4): # Thickness
        r = 50 + r_offset
        # i is row (y-down), j is col (x-right)
        # sin(theta) 0->1->0 gives positive offset (downwards)
        # cos(theta) 1->0->-1 gives right to left sweep
        x = int(cx + r*np.sin(theta)) 
        y = int(cy + r*np.cos(theta))
        if 0 <= x < size and 0 <= y < size:
            Z[x, y] = 0   # Smile (Black)

plt.imshow(Z, cmap='inferno')
plt.axis('off')
plt.show()
```
